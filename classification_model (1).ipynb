```python
#RESNET152
base_model=tf.keras.applications.resnet.ResNet152(
    include_top=False,
    weights='imagenet',
    input_shape=(81,81,3)
    #input_tensor=None
)
```

```python
# we don’t have to train all the layers, we make them non_trainable
for layer in base_model.layers:
    layer.trainable = False
    
# Flatten the output layer to 1 dimension
x = Flatten()(base_model.output)
#fully connected layer with 512 hidden units and ReLU activation
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)

x = Dense(19,activation='softmax')(x) 
model_resnet = tf.keras.models.Model(base_model.input,x)
model_resnet.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model_resnet.fit(X_train, Y_train, epochs = 27 ,validation_split=0.1)
```
